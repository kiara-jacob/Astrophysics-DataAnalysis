```python
data = pd.read_csv("m64.csv")
y=data["dec"]
x=data['ra']
res = sns.kdeplot(x,y,color='blue',shade=True)
plt.xlabel("RA")
plt.ylabel("DEC")
plt.title("M64: The Black Eye Galaxy")
plt.show()
```
